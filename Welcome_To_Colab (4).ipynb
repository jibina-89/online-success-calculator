```
import numpy as np
import IPython.display as display
from matplotlib import pyplot as plt
import io
import base64

ys = 200 + np.random.randn(100)
x = [x for x in range(len(ys))]

fig = plt.figure(figsize=(4, 3), facecolor='w')
plt.plot(x, ys, '-')
plt.fill_between(x, ys, 195, where=(ys > 195), facecolor='g', alpha=0.6)
plt.title("Sample Visualization", fontsize=10)

data = io.BytesIO()
plt.savefig(data)
image = F"data:image/png;base64,{base64.b64encode(data.getvalue()).decode()}"
alt = "Sample Visualization"
display.display(display.Markdown(F"""![{alt}]({image})"""))
plt.close(fig)
```

```
import requests
from bs4 import BeautifulSoup
import pandas as pd
import time

data = []

# Coursera course sitemaps (multiple files)
sitemaps = [
    "https://www.coursera.org/sitemap~www~courses.xml",
    "https://www.coursera.org/sitemap~www~courses~1.xml",
    "https://www.coursera.org/sitemap~www~courses~2.xml",
]

headers = {"User-Agent": "Mozilla/5.0"}

for sitemap_url in sitemaps:
    print("Reading sitemap:", sitemap_url)

    r = requests.get(sitemap_url, headers=headers)
    soup = BeautifulSoup(r.text, "xml")

    urls = soup.find_all("loc")

    for loc in urls:
        course_url = loc.text
        title = course_url.split("/")[-1].replace("-", " ").title()
        category = course_url.split("/")[4] if len(course_url.split("/")) > 4 else "General"

        data.append([title, "Coursera", category, course_url])

        if len(data) >= 3000:
            break

    if len(data) >= 3000:
        break

    time.sleep(1)

df = pd.DataFrame(
    data,
    columns=["Course Title", "Platform", "Category", "URL"]
)

df.to_csv("coursera_3000_courses.csv", index=False)

print("‚úÖ Dataset saved with", len(df), "rows")
```

```
import requests, json, time, re
import pandas as pd
from bs4 import BeautifulSoup

headers = {"User-Agent": "Mozilla/5.0"}
data = []

# Coursera sitemap
sitemap = "https://www.coursera.org/sitemap~www~courses.xml"

xml = requests.get(sitemap, headers=headers).text
soup = BeautifulSoup(xml, "xml")
course_urls = [loc.text for loc in soup.find_all("loc")][:3000]

for idx, url in enumerate(course_urls):
    print(f"{idx+1} Scraping:", url)

    try:
        html = requests.get(url, headers=headers, timeout=10).text
        soup = BeautifulSoup(html, "html.parser")

        # Extract JSON-LD
        script = soup.find("script", type="application/ld+json")
        if not script:
            continue

        meta = json.loads(script.string)

        title = meta.get("name")
        rating = meta.get("aggregateRating", {}).get("ratingValue")
        learners = meta.get("aggregateRating", {}).get("ratingCount")
        duration = meta.get("timeRequired")
        instructor = meta.get("creator", {}).get("name")
        platform = "Coursera"

        # Category from URL
        category = url.split("/")[4] if len(url.split("/")) > 4 else "General"

        # Difficulty (rule-based)
        difficulty = "Beginner"
        if duration and "PT20H" in duration:
            difficulty = "Intermediate"
        if duration and "PT40H" in duration:
            difficulty = "Advanced"

        # Price (derived ‚Äì Coursera standard)
        price = "Free / Paid"

        data.append([
            title, platform, category, price,
            duration, difficulty, instructor,
            rating, learners
        ])

    except:
        continue

    time.sleep(0.5)

df = pd.DataFrame(data, columns=[
    "Title", "Platform", "Category", "Price",
    "Duration", "Difficulty", "Instructor",
    "Rating", "Learners"
])

df.to_csv("courses_3000_full_attributes.csv", index=False)
print("‚úÖ Dataset saved with", len(df), "rows")
```

```
import requests
import pandas as pd
import json
import time
import re
import numpy as np
from bs4 import BeautifulSoup

headers = {"User-Agent": "Mozilla/5.0"}

# ---------------------------------------
# 1. GET 5000 COURSE URLS FROM SITEMAPS
# ---------------------------------------
sitemaps = [
    "https://www.coursera.org/sitemap~www~courses.xml",
    "https://www.coursera.org/sitemap~www~courses~1.xml",
    "https://www.coursera.org/sitemap~www~courses~2.xml"
]

course_urls = []

for sm in sitemaps:
    xml = requests.get(sm, headers=headers).text
    soup = BeautifulSoup(xml, "xml")
    course_urls.extend([loc.text for loc in soup.find_all("loc")])

course_urls = course_urls[:5000]
print("Collected URLs:", len(course_urls))

# ---------------------------------------
# 2. SCRAPE METADATA FROM COURSE PAGES
# ---------------------------------------
data = []

for i, url in enumerate(course_urls):
    print(f"{i+1}/5000")
    try:
        html = requests.get(url, headers=headers, timeout=10).text
        soup = BeautifulSoup(html, "html.parser")

        script = soup.find("script", type="application/ld+json")
        if not script:
            continue

        meta = json.loads(script.string)

        title = meta.get("name", "Unknown Course")
        instructor = meta.get("creator", {}).get("name", "Unknown")
        rating = meta.get("aggregateRating", {}).get("ratingValue")
        learners = meta.get("aggregateRating", {}).get("ratingCount")
        duration_raw = meta.get("timeRequired")

        # Convert PT54H ‚Üí 54
        duration_hours = None
        if duration_raw:
            m = re.search(r"PT(\d+)H", duration_raw)
            if m:
                duration_hours = int(m.group(1))

        category = url.split("/")[4] if len(url.split("/")) > 4 else "General"

        data.append([
            title,
            "Coursera",
            category,
            "Free / Paid",
            instructor,
            rating,
            learners,
            duration_hours
        ])

    except:
        pass

    time.sleep(0.25)

df = pd.DataFrame(data, columns=[
    "Title", "Platform", "Category", "Price",
    "Instructor", "Rating", "Learners", "Duration_Hours"
])

print("Raw rows:", len(df))

# ---------------------------------------
# 3. FILL ALL MISSING VALUES (MANDATORY)
# ---------------------------------------
df["Rating"] = pd.to_numeric(df["Rating"], errors="coerce")
df["Learners"] = pd.to_numeric(df["Learners"], errors="coerce")
df["Duration_Hours"] = pd.to_numeric(df["Duration_Hours"], errors="coerce")

# Rating ‚Üí global mean
df["Rating"].fillna(df["Rating"].mean(), inplace=True)

# Learners ‚Üí category mean, then median
df["Learners"] = df.groupby("Category")["Learners"].transform(
    lambda x: x.fillna(x.mean())
)
df["Learners"].fillna(df["Learners"].median(), inplace=True)

# Duration ‚Üí median
df["Duration_Hours"].fillna(df["Duration_Hours"].median(), inplace=True)

# Instructor
df["Instructor"].replace("", "Unknown", inplace=True)
df["Instructor"].fillna("Unknown", inplace=True)

# ---------------------------------------
# 4. DERIVE DIFFICULTY
# ---------------------------------------
def difficulty(hours):
    if hours <= 10:
        return "Beginner"
    elif hours <= 30:
        return "Intermediate"
    else:
        return "Advanced"

df["Difficulty"] = df["Duration_Hours"].apply(difficulty)

# ---------------------------------------
# 5. FINAL DATASET
# ---------------------------------------
df = df.drop_duplicates().reset_index(drop=True)
df.to_csv("courses_5000_full_attributes.csv", index=False)

print("‚úÖ FINAL DATASET SHAPE:", df.shape)
df.head()
```
